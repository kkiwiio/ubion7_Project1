```python
import pandas as pd
```

```python
dh = pd.read_excel('./datasets/0626/대한항공 주식 데이터.xlsx')
lg = pd.read_excel('./datasets/0626/lg화학 주식 데이터.xlsx')
```

```python
dh.columns = ['일자', '대한종가', '대한변동', '대한시가총액']
lg.columns = ['일자', '엘지종가', '엘지변동', '엘지시가총액']
```

```python
total = pd.merge(dh, lg, how="inner")
```

```python
total.sort_values(by="일자", ignore_index=True, inplace=True)
```

```python
total
```

```python
total["DL"] = (total["대한시가총액"] + total["엘지시가총액"])
```

```python
# total['DG'] = round(total['DG']*100, 2)
```

```python
total
```

```python
_standard = total["DL"][0]
```

```python
total["DL"] = (total["DL"] / _standard) * 100
```

```python
total['target']  = total['DL'].diff().apply(lambda x: 1 if x > 0 else 0)
```

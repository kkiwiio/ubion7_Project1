```python
import pandas as pd
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
news_df = pd.read_excel('./유가 전처리 (완) 인베딩 (미완).xlsx')
```

```python
news_df
```

```python
model_name = "beomi/kcbert-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
texts = news_df['oil_news'].tolist()
```

```python
encoded_inputs = tokenizer(texts, truncation=True, padding=True, return_tensors='pt')
with torch.no_grad():
    outputs = model(**encoded_inputs)
embeddings = outputs.last_hidden_state[:, 0, :].numpy()
news_df['embedding'] = embeddings.tolist()
```

```python
news_df.to_excel("./유가 전처리 (완) 인베딩 (완).xlsx")
```
